# matplotlib

python의 대표적인 시각화 도구

다양한 Graph를 지원해주고 Pandas와의 연동도 지원!

````python
import matplotlib.pyplot as plt

X = range(100)
Y = [value**2 for value in X]
plt.plot(X, Y) # pyplot 객체에 그래프를 쌓은 다음 show를 통해 그래프를 그림 (flush)
plt.show()
````

graph는 원래 figure 객체에 생성된다.

pyplot 객체를 사용하면 기본 `figure`에 그래프가 그려진다.

````python
import numpy as np

X_1 = range(100)
Y_1 = [np.cos(value) for value in X]

X_2 = range(100)
Y_2 = [np.sin(value) for value in X]

plt.plot(X_1, Y_1)
plt.plot(X_2, Y_2)
plt.show()
````

````python
fig = plt.figure()
fig.set_size_inches(10, 5) # 가로, 세로
ax_1 = fig.add_subplot(1, 2, 1)
ax_2 = fig.add_subplot(1, 2, 2)

# c = color
ax_1.plot(X_1, Y_1, c='b')
ax_2.plot(X_2, Y_2, c='g')
plt.show()
````

plot에서 color를 지정할 수 있다. 이를 color 인자를 통해 지정한다. (축약형으로 `c` 인자로도 지정할 수 있다. 위 color는 선의 색깔)

```python
plt.plot(X_1, Y_1, color='#eeefff')
```

color의 경우 몇몇 color에 대해서는 축약형을 제공해준다.

※ 참고
- 'b' : blue
- 'g' : green
- 'r' : red
- 'c' : cyan
- 'm' : magenta
- 'y' : yellow
- 'k' : black
- 'w' : white

이외에도 튜플형태로 (R, G, B) 값을 주거나 String 형태로 hex 값을 제공하여 color를 지정할 수 있다. color는 html에서 키워드로 등록된 color 이름을 사용할 수도 있다.

또한, linestyle / ls 속성을 통해서 linestyle을 지정할 수 있다.
```python
plt.plot(X_1, Y_1, linestyle='dashed')
```

선종류로는 기본적으로 `-'solid', --'dashed', :'dotted', -.'dashdotted'`가 있다.

plot의 title은 `plt.title()` 메서드를 통해 현재 plot의 제목을 달 수 있다.
> Latex 타입의 표현도 가능!

````python
X = range(100)
Y = range(100)
plt.plot(X, Y, linestyle=':')
plt.title('test', color='#a9a9a9')
plt.show()
````

그래프에서 자주 볼 수 있는 범례도 만들 수 있다.

plt의 legend 메서드를 통해 범례를 줄 수 있다. (단, 범례를 주기 위해서는 하나의 plot마다 label을 달아야한다.)
주로 shadow, fancybox, loc 인자를 사용한다. shadow는 범례에 그림자를 주기 위한 옵션, fancybox는 범례 box의 테두리를 둥글게 만들어주는 옵션, loc은 범례의 위치를 주는 옵션이다.

loc의 경우 `upper`, `lower`, `center`, `right`, `left`를 조합하여 위치를 조절할 수 있다.

x축과 y축의 범위를 줄 수도 있다.

xlim, ylim으로 범위를 줄 수 있다. 첫번째 인자는 최소 범위, 두번째 인자는 최대 범위가 된다.

````python
X = range(100)
Y = range(100)
plt.plot(X, Y, linestyle='-', label='line1')
plt.plot(X, Y, linestyle='-', color="#ff0000", label='line2')
plt.title('test', color='#a9a9a9')
plt.legend(shadow=True, fancybox=True, loc='upper center')
plt.grid(True, lw=0.5, ls='--', c='0.9')
plt.show()
````
